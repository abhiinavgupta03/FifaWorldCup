```python
fifa_rank=df[['date', 'home_team', 'away_team','home_team_fifa_rank', 'away_team_fifa_rank','home_team_total_fifa_points', 'away_team_total_fifa_points']]
home=fifa_rank[['date', 'home_team','home_team_fifa_rank','home_team_total_fifa_points']].rename(columns={'home_team':'team','home_team_fifa_rank':'rank','home_team_total_fifa_points':'rank_points'})
away=fifa_rank[['date', 'away_team','away_team_fifa_rank','away_team_total_fifa_points']].rename(columns={'away_team':'team','away_team_fifa_rank':'rank','away_team_total_fifa_points':'rank_points'})
fifa_rank = pd.concat([home, away])
fifa_rank=fifa_rank.sort_values(['team','date'],ascending=[True,False])
fifa_rank['row_number']=fifa_rank.groupby('team').cumcount()+1
fifa_rank_top=fifa_rank[fifa_rank['row_number']==1].drop('row_number',axis=1).nsmallest(10,'rank')
fifa_rank_top
```

```python
colums_contains_null=[col for col in df.columns if df[col].isnull().any()]
colums_contains_null
```

```python
for col in colums_contains_null:
    df[col]=df[col].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['away_team_code'] = df['away_team'].astype('category').cat.codes
df['day_code'] = df['date'].dt.dayofweek
```

```python
df.head()
```

```python
df.home_team_result
```

```python
df['neutral_loc']=(df['neutral_location']==True).astype('int')
df['penalties']=(df['shoot_out']=='Yes').astype('int')
df['target']=(df['home_team_result']=='Win').astype('int')
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
rf=RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=10)
```

```python
train=df[df['date']<'2022-01-01']
test=df[df['date']>'2022-01-01']
```

```python
df.columns
```

```python
predictors=['away_team_code', 'day_code', 'neutral_loc', 'penalties', 'home_team_goalkeeper_score', 'away_team_goalkeeper_score',
'home_team_mean_defense_score', 'home_team_mean_offense_score', 'home_team_mean_midfield_score', 'away_team_mean_defense_score',
'away_team_mean_offense_score', 'away_team_mean_midfield_score', 'home_team_fifa_rank', 'away_team_fifa_rank', 'home_team_total_fifa_points','away_team_total_fifa_points']
```

```python
rf.fit(train[predictors],train['target'])
```

```python
pred=rf.predict(test[predictors])
```

```python
pred[:10]
```

Testing accuray

```python
acc=accuracy_score(test['target'],pred)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test['target'], predcition=pred))
pd.crosstab(index=combined['actual'], columns=combined['predcition'])
```

```python
france_home = df[df['home_team'] == 'France'].sort_values(['date'], ascending = False)[:1]
france_away = df[df['away_team'] == 'France'].sort_values(['date'], ascending = False)[:1]
morocco_home = df[df['home_team'] == 'Morocco'].sort_values(['date'], ascending = False)[:1]
morocco_away=df[df['away_team'] == 'Morocco'].sort_values(['date'], ascending = False)[:1]
argentina_home = df[df['home_team'] == 'Argentina'].sort_values(['date'], ascending = False)[:1]
argentina_away = df[df['away_team'] == 'Argentina'].sort_values(['date'], ascending = False)[:1]
croatia_home =df[df['home_team'] == 'Croatia'].sort_values(['date'], ascending = False)[:1]
croatia_away = df[df['away_team'] == 'Croatia'].sort_values(['date'], ascending = False)[:1]
```

```python
semi_fin = pd.concat([france_home,france_away, morocco_home, morocco_away, argentina_home, argentina_away, croatia_home, croatia_away])
```

```python
semi_fin
```

```python
semi_fin['home_team_fifa_rank'] = [4,33,22,15,3,6, 16, 4]
semi_fin['home_team_total_fifa_points'] = [1759,1500,1563,1633,1765,1773,1621,1759]
semi_fin['away_team_fifa_rank'] = [16,4,149,22,110,3,4,16]
semi_fin['away_team_total_fifa_points'] = [1621,1759,1050,1563,1169,1765,1759,1621]
semi_fin
```

```python
home_stats = semi_fin.filter(regex='home')
away_stats = semi_fin.filter(regex='away')
```

```python
final_home = home_stats[(home_stats['home_team'] == 'Argentina') | (home_stats['home_team'] == 'France')]
final_away = away_stats[(away_stats['away_team'] == 'Argentina') | (away_stats['away_team'] == 'France')]
```

```python
final_home = final_home[:-1]
final_away = final_away[:-1]
```

```python
dates= [pd.to_datetime('2022-12-18'), pd.to_datetime('2022-12-18')]

final_home.insert(loc=0,column='date', value=dates)
final_home['day_code'] = [6,6]
final_home['neutral_loc'] = [1,1]
final_home['penalties'] = [0,0]
```

```python
final_home.index =[0,1]
final_away.index =[1,0]
final_away
```

```python
finals = pd.concat([final_home, final_away],axis=1)
```
